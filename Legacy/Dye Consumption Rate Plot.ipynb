# Calculation for the Dye Consumption Rate

### Note: With a UV-Vis or spectrofluorimeter, we are in fact measuring the amount of unbleached dye in the system. That is not exactly what we want; the equation we have solved is for the consumed bleach dye - so that's what we need. But since this is a quasi-steadystate measurement, i.e. the measurements are taken at discrete intervals where the reaction has gone to completion each time, it is safe to assume that the bleached dye = 1 - unbleached dye.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Example user data (replace with experimental values)
time = np.array([0, 0.5, 1, 2, 3]) # seconds
intensity = np.array([1.0, 0.9, 0.5, 0.3, 0.2]) # arbirary units normalized
Bleached_Dye = 1 - intensity

# fit curve
result = linregress(time, Bleached_Dye)
dBdt = result.slope

print(f"Rate of Dye Consumption: {dBdt:.3e} s^-1")

# Plot dBdt vs t
plt.figure()
plt.scatter(time, Bleached_Dye, color='blue', label='Data')
plt.plot(time, result.intercept + result.slope * time, color='red', label=f'Fit: k = {dBdt:.2e} s⁻¹')
plt.xlabel('Time (s)')
plt.ylabel('Remaining Dye in system [B]')
plt.title('Dye Consumption Rate Determination')
plt.legend()
plt.grid(True)
plt.show()
```
